```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt



from sklearn.metrics import silhouette_score
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
df=df.drop('CustomerID',axis=1)
```

```python
mp={'Male':0 , 'Female':1}
df['Genre']=df['Genre'].map(mp)


```

```python
df.head()
```

```python
wclussum=[]
from sklearn.cluster import KMeans

for i in range(1,15):
  km=KMeans(n_clusters=i)
  km.fit(df)
  wclussum.append(km.inertia_)



```

```python
print(wclussum)

plt.plot(range(1,15), wclussum)
plt.title('Inertia')
plt.xlabel('Clusters')
plt.ylabel('within Cluster Sum Square')
plt.show()
```

```python
kmn=KMeans(n_clusters=6)

kmn.fit(df)
```

```python

silhouette_score(df,kmn.predict(df))
```
